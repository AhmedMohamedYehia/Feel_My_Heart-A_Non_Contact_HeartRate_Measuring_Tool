## Harun Resting

<hr>

```python
harun_no_detrending = pd.read_csv(r'D:\Uni\GP\Feel_My_Heart-A_Non_Contact_HeartRate_Measuring_Tool\Testing\Harun_Resting\No_Detrending\HR_Results.csv',header=None)
harun_detrended = pd.read_csv(r'D:\Uni\GP\Feel_My_Heart-A_Non_Contact_HeartRate_Measuring_Tool\Testing\Harun_Resting\Detrended\HR_Results.csv',header=None)
harun_Old_Approach = pd.read_csv(r'D:\Uni\GP\Feel_My_Heart-A_Non_Contact_HeartRate_Measuring_Tool\Testing\Harun_Resting\Old_Approach\HR_Results.csv',header=None)
harun_Ground_Truth = pd.read_csv(r'D:\Uni\GP\Feel_My_Heart-A_Non_Contact_HeartRate_Measuring_Tool\Testing\Harun_Resting\Ground_Truth\HR_Ground_Truth.csv',header=None)


print("harun_Old_Approach: ",harun_Old_Approach)
print("harun_no_detrending: ",harun_no_detrending)
print("harun_Alexdetrended: ",harun_detrended)
print("harun_Ground_Truth: ",harun_Ground_Truth)
```

```python
print("RMSLE:")
print("harun_Old_Approach: ",np.sqrt(mean_squared_log_error(harun_Ground_Truth, harun_Old_Approach)))
print("harun_detrended: ",np.sqrt(mean_squared_log_error(harun_Ground_Truth, harun_detrended)))
print("harun_no_detrending: ",np.sqrt(mean_squared_log_error(harun_Ground_Truth, harun_no_detrending)))
```

```python

fig = plt.figure(figsize=(16, 8))
fig.patch.set_facecolor('white')
plt.plot(harun_Ground_Truth)
plt.plot(harun_Old_Approach)
plt.legend(["Ground Truth","Old Approach"])

plt.xlabel('Frame Index', fontsize=17)
plt.ylabel("Heart Rate", fontsize=17)
plt.ylim([45,180])
# plt.savefig(r'D:\Uni\GP\testing_figures\1.1.png')
plt.show()
```

```python

fig = plt.figure(figsize=(16, 8))
fig.patch.set_facecolor('white')

plt.plot(harun_Ground_Truth)
plt.plot(harun_detrended)
plt.plot(harun_no_detrending)

plt.xlabel('Frame Index', fontsize=17)
plt.ylabel("Heart Rate", fontsize=17)
plt.ylim([45,150])
plt.legend(["Ground Truth","Detrended","Not Detrended"])
plt.savefig(r'D:\Uni\GP\testing_figures\4.1.png')
plt.show()
```
